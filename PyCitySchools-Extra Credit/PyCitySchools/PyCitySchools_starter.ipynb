```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# For District Summary we are calculating:

# Total number of schools
schools_names_df = school_data_complete['school_name'].unique()
total_num_schools = len(schools_names_df)
total_num_schools

# Total number of students
total_num_students = school_data_complete['student_name'].count()
total_num_students

# Total budget
total_budget = (school_data_complete['budget'].unique()).sum()
total_budget

# Average math score
math_avg_score = school_data_complete['math_score'].mean()
math_avg_score

# Average reading score
reading_avg_score = school_data_complete['reading_score'].mean()
reading_avg_score

# Percentage of students with a passing math score (70 or >) using the loc function for filter by passing the condition (>=70)
math_pass = school_data_complete.loc[school_data_complete['math_score'] >= 70]['math_score'].count()

percentage_math = math_pass/total_num_students
percentage_math

# Percentage of students with a passing reading score (70 or >)
reading_pass = school_data_complete.loc[school_data_complete['reading_score'] >= 70]['reading_score'].count()

percentage_reading = reading_pass/total_num_students
percentage_reading

# Percentage of students who passed math and reading (% Overall Passing)

passing_overall = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['Student ID'].count()/total_num_students
passing_overall



# Creating new dataframe
district_summary = pd.DataFrame({
    "Total Schools": [total_num_schools],
    "Total Students": [total_num_students],
    "Total Budget": [total_budget],
    "Average Math Score": [math_avg_score],
    "Average Reading Score": [reading_avg_score],
    "% Passing Math": [percentage_math],
    "% Passing Reading": [percentage_reading],
    "% Overall Passing": [passing_overall]
})

# district_summary["Total Students"] = district_summary["Total Students"].map("{:,.2}".format)


district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:

* Create a dataframe to hold the above results

```python
# For School Summary: 

# School name using set_index & groupBy for school names
schools_names_df = school_data_complete.set_index("school_name").groupby(["school_name"])

# School Type
school_type = school_data.set_index("school_name")["type"]
school_type

# Total Students 
total_students = schools_names_df["Student ID"].count()
total_students

# Total School Budget
total_school_budget = school_data.set_index("school_name")["budget"]
total_school_budget

# Per Student Budget
student_budget = total_school_budget / (school_data.set_index("school_name")["size"])
student_budget

# Average Math Score
avg_math_score = schools_names_df["math_score"].mean()

# Average Reading Score
avg_reading_score = schools_names_df["reading_score"].mean()
```

```python
# Percentage of Passing Math (using groupby to group math score by student name using student ID)
passing_math_per = school_data_complete[school_data_complete['math_score'] >= 70].groupby("school_name")["Student ID"].count()/total_students


# Percentage of Passing Reading (using groupby to group reading score by student name using student ID)
passing_reading_per = school_data_complete[school_data_complete['reading_score'] >= 70].groupby("school_name")["Student ID"].count()/total_students


# Percentage of Overall Passing (The percentage of students that passed math and reading.)

overall_passing_per = school_data_complete[(school_data_complete['reading_score'] >= 70) & 
                        (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_students

```

```python
# I was getting this (ValueError: cannot reindex from a duplicate axis) so I seperated the dataframe from calculations


# Creating new dataframe for school summary
school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_students,
    "Total School Budget": total_school_budget,
    "Per Student Budget": student_budget,
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_reading_score,
    "% Passing Math": passing_math_per,
    "% Passing Reading": passing_reading_per,
    "% Overall Passing": overall_passing_per
})

# Formatting School Summary new dataframe

school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,}".format)
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:.2f}".format)
school_summary_df["Average Math Score"] = school_summary_df["Average Math Score"].map("{:.1f}".format)
school_summary_df["Average Reading Score"] = school_summary_df["Average Reading Score"].map("{:.1f}".format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map("{:.2%}".format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("{:.2%}".format)
school_summary_df["% Overall Passing"] = school_summary_df["% Overall Passing"].map("{:.2%}".format)

school_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Creating bins for school scores by spending: 

bins = [0, 584, 629, 644, 675]

group_names = ["<$584", "$585-629", "$630-644", "$645-675"]

school_data_complete["spending_range"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels=group_names)

# Creating a group based off spending

spending_bins = school_data_complete.groupby("spending_range")
```

```python
# Calculating math average scores by spending

math_avg_spending = spending_bins['math_score'].mean()

# Calculation reading average scores by spending 

reading_avg_spending = spending_bins['reading_score'].mean()
```

```python
# Calculating the percentage of passing math 

passing_math = school_data_complete[school_data_complete["math_score"] >=70].groupby("spending_range")["Student ID"].count()/spending_bins["Student ID"].count()*100

# Calculating the percentage of passing reading

passing_reading = school_data_complete[school_data_complete["reading_score"] >=70].groupby("spending_range")["Student ID"].count()/spending_bins["Student ID"].count()*100

# Overall Passing Rate (Average of the above two) using '&' in groupby

overall_passing = school_data_complete[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"] >=70)].groupby("spending_range")["Student ID"].count()/spending_bins["Student ID"].count()*100 
```

```python
# Creating new dataframe for scores by school spending 

scores_spending_df = pd.DataFrame({
    "Average Math Score": math_avg_spending,
    "Average Reading Score": reading_avg_spending,
    "% Passing Math": passing_math,
    "% Passing Reading": passing_reading,
    "% Overall Passing": overall_passing
})

scores_spending_df

```

```python
# Seperated the formatting as it was giving me errors

# Formatting Data
scores_spending_df.index.name = "Spending Ranges(Per Student)"
scores_spending_df = scores_spending_df.reindex(group_names)


scores_spending_df["Average Math Score"] = scores_spending_df["Average Math Score"].map("{:.1f}".format)
scores_spending_df["Average Reading Score"] = scores_spending_df["Average Reading Score"].map("{:.1f}".format)
scores_spending_df["% Passing Math"] = scores_spending_df["% Passing Math"].map("{:.2f}".format)
scores_spending_df["% Passing Reading"] = scores_spending_df["% Passing Reading"].map("{:.2f}".format)
scores_spending_df["% Overall Passing"] = scores_spending_df["% Overall Passing"].map("{:.2f}".format)

scores_spending_df
```
